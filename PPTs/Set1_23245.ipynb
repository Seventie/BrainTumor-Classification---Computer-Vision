```python
import  requests                        #html page fetch
from  bs4  import BeautifulSoup
import re                       #regular expression
import string                #handling strings
import nltk                 #natural language toolkit
from  nltk.corpus   import  stopwords       #Data representation
from nltk.probability import FreqDist             #Frequency Distribution
from nltk.util    import ngrams      #ngrams
import matplotlib.pyplot   as plt    #Visualization
```

```python
b.	Scrap data from https://intranet.cb.amrita.edu/events/2025/Gokulashtami-2025
and display.                                                                                   [2][CO01][BTL2]
```

```python
#add code
```

```python
# Step 1: Scrape data from the webpage
url = "https://intranet.cb.amrita.edu/events/2025/Gokulashtami-2025"
try:
    response = requests.get(url)
    html_content = response.text
    print("Webpage content:\n", html_content)   # directly print content
except Exception as e:
    print(f"Failed to retrieve data: {e}")
    html_content = ""
```

c.	                                                                                                     [3] [CO01][BTL2]

i.	Remove all html tags and store the text.

ii.	Convert into lower case, remove punctuation and numbers, and store the text.


```python
# Step 2: Remove HTML tags and store the text
soup = BeautifulSoup(html_content, "html.parser")
text = soup.get_text(separator=" ", strip=True)
print("After removing HTML tags:\n", text[:500], "...")

# Step 3: Preprocess the text
# i. Convert into lowercase
text_clean = text.lower()

# ii. Remove punctuation
text_clean = re.sub(r"[{}]".format(string.punctuation), "", text_clean)

# iii. Remove numbers
text_clean = re.sub(r"[0-9]", "", text_clean)

print("\nAfter preprocessing:\n", text_clean[:500], "...")
```

d.Display lines beginning with  Re and ending with ts using regular expressions.
                                                                                                          [5 ][CO02][BTL3]

```python
# Step 4: Display lines beginning with 'Solo' and ending with 'ion'
pattern_lines = re.findall(r"^Re.*ts$", text_clean, re.MULTILINE)

print("\nLines beginning with 'Re' and ending with 'ts':")
for line in pattern_lines:
    print(line)

```

e.	Compute TF-IDF for the stored text and display .                           [5 ][CO02][BTL3]


```python
# Step 5: Compute TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([text_clean])  
feature_names = vectorizer.get_feature_names_out()      
dense = tfidf_matrix.todense()                          
tfidf_scores = dense[0].tolist()[0]                    

print("\nTF-IDF Scores:")
for word, score in zip(feature_names, tfidf_scores):
    print(f"{word}: {score:.4f}")
```

f.	Display the frequency distribution of words that begin and end with the same letter.
    e.g. amrita                                                                                         [5] [CO02][BTL3]
                                                                                                    

```python
words = nltk.word_tokenize(text_clean)
filtered_words = [w for w in words if len(w) > 1 and w[0] == w[-1]]
fdist = nltk.FreqDist(filtered_words)

print("\nFrequency distribution of words starting and ending with the same letter:")
for word, freq in fdist.items():
    print(f"{word}: {freq}")
```

Qn 6.  [5] [CO03][BTL4]

                                            
a.	Find the trigrams

```python
trigrams = list(ngrams(words,3))
print("trigrams",trigrams[:10])
```

b. Fetch the ten least frequently used trigrams 


```python
trigram_freq = FreqDist(trigrams)
least_common_trigrams = trigram_freq.most_common()[:-6:-1]

print("\nLeast common trigrams:")
for trigram, freq in least_common_trigrams:
    print(f"{trigram}: {freq}")
```

c.Visualize the same.

```python
# Step 8: Visualize least frequent bigrams and trigrams
def plot_ngrams(ngrams_list, title):
    labels = [f"{' '.join(ngram)}" for ngram, _ in ngrams_list]
    values = [freq for _, freq in ngrams_list]
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values)
    plt.ylabel("Frequency")
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
plot_ngrams(least_common_trigrams, "Least Frequent trigrams")

```
